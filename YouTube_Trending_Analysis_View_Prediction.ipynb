```python
X1 = Trend_Video_US_CAT[[ 'category_id','likes', 'dislikes', 'comment_count']]
Y1=Trend_Video_US_CAT.views
```

```python
scalar=MinMaxScaler()
```

```python
X1_train, X1_test, Y1_train, Y1_test = train_test_split(X1, Y1, test_size = .2)
```

```python
scalar.fit(X1_train)
X_train_scaled = scalar.transform(X1_train)
X_test_scaled = scalar.transform(X1_test)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
regr=linear_model.LinearRegression()
```

```python
regr.fit(X_train_scaled, Y1_train)
```

```python
y_pred = regr.predict(X_test_scaled)
```

```python
print("Coefficient of determination: %.2f" % r2_score(Y1_test, y_pred))
```
